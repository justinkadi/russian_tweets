```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Uploading tweets.csv that contains the text and information about the actual tweets
tweets = pd.read_csv('russian-troll-tweets/tweets.csv')
tweets_df = pd.DataFrame(tweets)
print('Shape: This set of data has', tweets_df.shape[0], 'tweets and', tweets_df.shape[1], 'columns to analyze.')
tweets_df.head()
```

```python
#I selected only some columns and assigned "tweet_id" to the index of the DataFrame
tdf = tweets_df[['user_key', 'created_str', 'retweet_count', 'favorite_count', 'text', 'tweet_id', 'hashtags', 'mentions']]
tdf = tdf.set_index('tweet_id')
tdf.head()
```

```python
tdf = tdf.dropna(subset=['text']) #Dropped rows with NaN text
tdf['created_str'] = pd.to_datetime(tdf['created_str'])
tdf = tdf.rename(index=str, columns={"created_str": "created_at"})
tdf.head()
```

```python
print(len(tdf['user_key'].unique()), "unique user keys")
```

```python
#source: https://stackoverflow.com/questions/6451655/python-how-to-convert-datetime-dates-to-decimal-years
import datetime
def year_fraction(date):
    start = datetime.date(date.year, 1, 1).toordinal()
    year_length = datetime.date(date.year+1, 1, 1).toordinal() - start
    return date.year + float(date.toordinal() - start) / year_length

tdf['year'] = tdf['created_at'].apply(year_fraction)
tdf.head()
```

```python
plt.figure(figsize=(8, 6))

sns.distplot(tdf['year'], label = 'year')
plt.title('Distributions of Tweets Over Years 2014-2018')
plt.xticks(np.arange(2014, 2019, 1));
```

```python
temp = tdf[(tdf['year'] < 2017)]
tweets_election_period = temp[temp['year'] > 2016]

plt.figure(figsize=(8, 6))

sns.distplot(tweets_election_period['year'], label = 'election')
plt.title('Distributions of Tweets Over Years 2016-2017')
plt.xticks(np.arange(2016, 2018, 1));
```
